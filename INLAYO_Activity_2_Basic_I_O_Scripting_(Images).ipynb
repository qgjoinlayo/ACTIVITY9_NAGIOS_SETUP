````python
import cv2
import matplotlib.pyplot as plt

# Path to your image in Google Drive
img_path = '/content/drive/MyDrive/Colab Notebooks/Yasuo_0.jpg'

# Read and convert to RGB
img_bgr = cv2.imread(img_path, cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Display
plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.axis('off')
plt.title('Yasuo_0.jpg from Google Drive')
plt.show()
````

````python
gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
cv2.imwrite('/content/drive/My Drive/images/Yasuo_grayscale.png', gray)

plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.title('Grayscale Yasuo')
plt.show()
````

````python
flags = {
    'IMREAD_UNCHANGED': cv2.IMREAD_UNCHANGED,
    'IMREAD_GRAYSCALE': cv2.IMREAD_GRAYSCALE,
    'IMREAD_COLOR': cv2.IMREAD_COLOR,
    'IMREAD_ANYDEPTH': cv2.IMREAD_ANYDEPTH,
    'IMREAD_ANYCOLOR': cv2.IMREAD_ANYCOLOR,
    'IMREAD_REDUCED_COLOR_2': cv2.IMREAD_REDUCED_COLOR_2,
}

observations = {}
for name, flag in flags.items():
    img = cv2.imread(img_path, flag)
    if img is None:
        observations[name] = 'Failed to load.'
        continue

    shape = img.shape
    dtype = img.dtype
    channels = 1 if len(shape) == 2 else shape[2]

    disp = img if channels == 1 else cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(5,5))
    plt.imshow(disp, cmap='gray' if channels == 1 else None)
    plt.axis('off')
    plt.title(name)
    plt.show()

    observations[name] = f"shape={shape}, dtype={dtype}, channels={channels}"

observations
````

````python
# Load image from Google Drive
img_path = '/content/drive/MyDrive/Colab Notebooks/Yasuo_0.jpg'
img_bgr = cv2.imread(img_path, cv2.IMREAD_COLOR)

# Final crop values based on image shape (717 x 1215)
y_start, y_end = 200, 400   # vertical range (rows)
x_start, x_end = 800, 1100   # horizontal range (columns)

roi_bgr = img_bgr[y_start:y_end, x_start:x_end]
roi_rgb = cv2.cvtColor(roi_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(roi_rgb)
plt.axis('off')
plt.title('Face ROI (Final Manual Crop)')
plt.show()
````
